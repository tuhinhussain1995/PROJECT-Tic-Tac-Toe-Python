```python
from IPython.display import clear_output

board = ['#', '1', '2', '3', '4', '5', '6', '7', '8', '9']
player = ['$', '&']
options = []

def display_board():
    global board
    
    print('   |   |   ')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |   ')
    print('-----------')
    print('   |   |   ')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |   ')
    print('-----------')
    print('   |   |   ')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])


def play_again():
    global board
    global player
    global options
    
    board = ['#', '1', '2', '3', '4', '5', '6', '7', '8', '9']
    player = ['$', '&']
    options = []
    
    
def player_input():
    global player
    
    while True:
        myPlayer = input("Enter First Player choice X or O: ")
        
        if myPlayer == 'X' or myPlayer == 'O':
            player[0] = myPlayer
            break
        
    if player[0] == 'X':
        player[1] = 'O'
    else:    
        player[1] = 'X'
        
    print(player)    
        
        
    
def player_choice(plr):
    global board
    
    while True:
        choice = input('Player {0} choose your next position: '.format(plr))
        
        if choice in board[1:]:
            board[int(choice)] = plr
            options.append(int(choice))
            break
    
    
def win_check(mark):
    global board

    return ((board[7] ==  board[8] ==  board[9] == mark) or 
            (board[4] ==  board[5] ==  board[6] == mark) or
            (board[1] ==  board[2] ==  board[3] == mark) or
            (board[7] ==  board[4] ==  board[1] == mark) or
            (board[8] ==  board[5] ==  board[2] == mark) or
            (board[9] ==  board[6] ==  board[3] == mark) or
            (board[7] ==  board[5] ==  board[3] == mark) or
            (board[9] ==  board[5] ==  board[1] == mark))   
    
    
    
def player_option(currentPlayer, process):
    global board
    global player
    global options
    
    while True:
        clear_output()

        display_board()

        player_choice(currentPlayer)

        if win_check(currentPlayer):
            clear_output()
            display_board()
            
            print("\nCongratulation player {0} wins!!!".format(currentPlayer))
            break

        if len(options) >= 9:
            clear_output()
            display_board()
            
            print("\nThe Game Is Draw!!!")
            break
            
        if process%2==0:
            currentPlayer = player[1]
            process = 1
        else:
            currentPlayer = player[0]
            process = 2
            

if __name__=='__main__':
    
    while True:
        display_board()
        player_input()
        
        currentPlayer = player[0]
        process = 2
        
        player_option(currentPlayer, process)
        
        decision = input("Do you want to play again Yes or No: ").
        
        if decision == 'No':
            break
            
        play_again()
        clear_output()
```
